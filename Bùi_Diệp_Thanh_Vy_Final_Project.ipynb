# EDA Exploratory Data Analysis

```python
#Xem mô tả các cột dữ liệu có type là int
statistic=df.describe()
statistic
```

```python
#Xem mô tả các cột dữ liệu có type là int
statistic=df.describe()
statistic.T
```

```python
#Tìm độ lệch của dữ liệu
df_skew = pd.DataFrame({'Column_name': pd.Series(dtype='str'),
                    'Skew': pd.Series(dtype='str')})
for cols in [column for column in statistic]:
  if statistic[cols].iloc[1]<statistic[cols].iloc[4]: #mean<median
    df_skew.loc[len(df_skew.index)]=[cols,'Negative']
  else:                                               #mean>median
    df_skew.loc[len(df_skew.index)]=[cols,'Positive']
df_skew
```

## EDA Univariate Analysis

Thống kê sự phân bố giá trị của các cột có loại dữ liệu là int

```python
#Tạo danh sách các cột cần thống kê
num_cols = df.select_dtypes(include=np.number).columns.tolist()
num_cols[0:12]
```

```python
#Vẽ biểu đồ thể hiện sự phân bố giá trị các cột có loại dữ liệu là int
for cols in num_cols:
  print(cols)
  print('Skew: ',round(df[cols].skew(),2))
  plt.figure(figsize=(15,4))
  plt.subplot(1,2,1)
  plt.hist(df[cols])
  plt.ylabel('count')
  plt.subplot(1,2,2)
  sns.boxplot(x=df[cols])
  plt.show()
```

# Advertisers' Profit

```python
# Tạo bảng summary cho các nhà quảng cáo
df_summary = pd.DataFrame({'Advertiser': pd.Series(dtype='str'),
                    'Total_clicks': pd.Series(dtype='float'),
                    'Total_bookings': pd.Series(dtype='float'),
                    'Total_cost': pd.Series(dtype='float'),
                    'Total_revenue': pd.Series(dtype='float')})
df_summary.loc[len(df_summary.index)]=['A',df['clicks_A'].sum(),df['bookings_A'].sum(),df['cost_A'].sum(),df['booking_rev_A'].sum()]
df_summary.loc[len(df_summary.index)]=['B',df['clicks_B'].sum(),df['bookings_B'].sum(),df['cost_B'].sum(),df['booking_rev_B'].sum()]
df_summary.loc[len(df_summary.index)]=['C',df['clicks_C'].sum(),df['bookings_C'].sum(),df['cost_C'].sum(),df['booking_rev_C'].sum()]
df_summary['Total_profit']=df_summary['Total_revenue']*0.15
df_summary
```

```python
#Xuất file csv để import vào powerbi
df_summary.to_csv('/content/drive/MyDrive/Colab Notebooks/BI17_PYTHON/Final Project/summary.csv',index=False)
```

```python
#Tạo bảng date từ cột date của dữ liệu
date = df.date.str.split('-', expand =True)
date
df['month']=date[1]
```

```python
df5=pd.pivot_table(df, index='month', values=['clicks_C','bookings_C'], aggfunc=np.sum, fill_value=0).reset_index()
df5
df5['booking_change_rate(%)']=0
for i in range(1,12):
  df5['booking_change_rate(%)'].iloc[i]=round((df5['bookings_C'].iloc[i]-df5['bookings_C'].iloc[i-1])/df5['bookings_C'].iloc[i-1]*100,0)
df5['click_change_rate(%)']=0
for i in range(1,12):
      df5['click_change_rate(%)'].iloc[i]=round((df5['clicks_C'].iloc[i]-df5['clicks_C'].iloc[i-1])/df5['clicks_C'].iloc[i-1]*100,0)

df5
```
